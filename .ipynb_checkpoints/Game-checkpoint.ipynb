```python
import cv2
import numpy as np
import os
import import_ipynb
import random
import cvzone
from hand_Track_Module import handDetector

cam_width, cam_height = 1280, 720 #攝像頭畫面寬&高大小設定
cap = cv2.VideoCapture(0) #攝像頭設置
cap.set(3, cam_width) #設定攝像頭畫面寬度(id為3表示寬)
cap.set(4, cam_height) #設定攝像頭畫面高度(id為4表示高)

backgroundPath = "Background/Grass Background.jpg" #存放背景圖片資料的路徑
garbagePath = "assets/P_garbage.png" #垃圾圖片的路徑
foodPath = "assets/P_food.png" #食物圖片的路徑
catPath = "assets/P_cat.png" #貓圖片的路徑

imgGarbage = cv2.imread(garbagePath, cv2.IMREAD_UNCHANGED) #讀取垃圾圖片
imgGarbage = cv2.resize(imgGarbage, (0,0), None, 0.25, 0.25) #設定垃圾圖片大小
imgFood = cv2.imread(foodPath, cv2.IMREAD_UNCHANGED) #讀取食物圖片
imgFood = cv2.resize(imgFood, (0,0), None, 0.25, 0.25) #設定食物圖片大小
imgCat = cv2.imread(catPath, cv2.IMREAD_UNCHANGED) #讀取貓圖片
imgCat = cv2.resize(imgCat, (0,0), None, 0.25, 0.25) #設定貓圖片大小

garbage = [] #垃圾位置存取陣列
food = [] #食物位置存取陣列
cat = 5,5 #貓的位置(起始位置)

catched = False #抓到物件的狀態
gameset = False #遊戲準備的狀態(食物&垃圾物件生成完畢後變為True，物件拉回起點後變為False)
gamestart = False #遊戲開始的狀態(當游標移到起始位置變為True，游標碰到物件後變為False)



#攝像頭出現遊玩畫面
def webcamPlayArea(gameSize_x, gameSize_y):
    offset_x, offset_y = int((cam_width - gameSize_x) / 2), int((cam_height - gameSize_y) / 2)
    limitMin_x, limitMin_y = offset_x, offset_y
    limitMax_x, limitMax_y = cam_width-offset_x, cam_height-offset_y 
    cv2.rectangle(img, (limitMin_x, limitMin_y), (limitMax_x, limitMax_y), (255,0,0), 2)
    
    
#垃圾&食物的座標生成
def objectsCoordinates(garbage, food, gameset):
    overlap=0
    garbage=[] #所有垃圾的座標位置清空
    for i in range(4):
        x=random.randint(100, 540)
        y=random.randint(50, 290)
        garbage.append([i,x,y]) 
            
    while overlap==0:
        x=random.randint(100, 540)
        y=random.randint(50, 290)
        overlap=1
        food=[x,y]
        for i in range(4):
            if x==garbage[i][1]:
                if y==garbage[i][2]:
                    overlap=0
    gameset = True
    return (garbage, food, gameset)    
    
    
#偵測碰撞
def detectCollision(gamePoint, gamestart, gameset):
    if gamePoint[0]>=cat[0]-5 and gamePoint[0]<=cat[0]+5:
        if gamePoint[1]>=cat[1]-5 and gamePoint[1]<=cat[1]+5:
            gamestart=True
    if gamestart==True:
        for i in range(4):
            if gamePoint[0]>=(garbage[i][1])-5 and gamePoint[0]<=(garbage[i][1])+5:
                if gamePoint[1]>=(garbage[i][2])-3 and gamePoint[1]<=(garbage[i][2])+3:
                    gameset=False
                    gamestart=False
    return (gamestart, gameset)


#物件出現在遊戲畫面
def objectsAppear(imgBackground, garbage, food, cat):
    imgBackground = cvzone.overlayPNG(imgBackground, imgCat, (cat[0], cat[1]))
    imgBackground = cvzone.overlayPNG(imgBackground, imgFood, (food[0], food[1]))
    for i in range(4):
        imgBackground = cvzone.overlayPNG(imgBackground, imgGarbage, (garbage[i][1], garbage[i][2]))
    return imgBackground


detector = handDetector() #建立偵測class


while True:
    #第一步驟
    retval, img = cap.read() #讀取攝像頭照到的影像，retval表示影像捕獲成功or失敗
    img = detector.findHands(img) #偵測手掌
    landmarkList = detector.findPosition(img, 0) #獲取手掌偵測節點資料
    img = np.flip(img, axis=1) #攝像頭畫面左右翻轉(axis=1表示左右翻轉，axis=0表示上下翻轉)
    img = np.ascontiguousarray(img)
    imgBackground = cv2.imread(backgroundPath)
    
    
    #第二步驟
    gameSize_x, gameSize_y = imgBackground.shape[1], imgBackground.shape[0] #獲取背景圖片大小(作為遊玩畫面大小)
    webcamPlayArea(gameSize_x, gameSize_y) #攝像頭繪製遊玩範圍
    
    
    #第三步驟
    if gameset==False:
        garbage, food, gameset = objectsCoordinates(garbage, food, gameset) #生成物件座標並改變遊戲準備狀態
    
    
    #第四步驟
    if len(landmarkList) != 0: #如果有偵測到手部的時候
        finger_x, finger_y = -(landmarkList[8][1] - cam_width), landmarkList[8][2] #獲取攝像頭的食指座標
        gamePoint = (finger_x - limitMin_x, finger_y - limitMin_y) #將攝像頭的食指座標轉換為遊戲座標
        
        if finger_x > limitMin_x and finger_x < limitMax_x and finger_y > limitMin_y and finger_y < limitMax_y: #若食指座標有在遊戲範圍內
            cv2.circle(imgBackground, gamePoint, 5, (0, 0, 255), cv2.FILLED) #在遊戲畫面顯示座標
        
        gamestart, gameset = detectCollision(gamePoint, gamestart, gameset) #偵測物體碰撞，並改變遊戲開始&準備的狀態
                        
    
    #第五步驟
    imgBackground = objectsAppear(imgBackground, garbage, food, cat)
    
    
    #第六步驟
    cv2.namedWindow("Image", cv2.WINDOW_NORMAL) #新增攝像頭視窗(Image)，允許視窗自由縮放
    cv2.resizeWindow("Image", img.shape[1]//2, img.shape[0]//2) #改變"Image"視窗大小
    cv2.imshow("Image", img) #攝像頭照到的影像顯示到"Image"視窗
    
    cv2.namedWindow("Game", cv2.WINDOW_NORMAL) #新增遊戲視窗(Game)，允許視窗自由縮放
    cv2.resizeWindow("Game", imgBackground.shape[1]*2, imgBackground.shape[0]*2) #改變"Game"視窗大小
    cv2.imshow("Game", imgBackground) #將背景圖片顯示到"Game"視窗
    
    key = cv2.waitKey(1) #等待1毫秒接收按鍵
    if key == 27: #若按下Esc按鍵
        break #退出迴圈
    
cap.release() #結束攝像機
cv2.destroyAllWindows() #關閉所有視窗
```
