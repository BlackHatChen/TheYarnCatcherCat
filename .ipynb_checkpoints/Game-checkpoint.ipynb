```python
import cv2
import numpy as np
import os
import import_ipynb
import random
import cvzone
from hand_Track_Module import handDetector

cam_width, cam_height = 1280, 720 #攝像頭畫面寬&高大小設定
cap = cv2.VideoCapture(0) #攝像頭設置
cap.set(3, cam_width) #設定攝像頭畫面寬度(id為3表示寬)
cap.set(4, cam_height) #設定攝像頭畫面高度(id為4表示高)

backgroundPath = "Background" #存放背景圖片資料的路徑
pathBackgrounds = os.listdir(backgroundPath) #取得背景來源
# print(pathBackgrounds) #測試有無讀取到資料夾中的內容

backgroundNumber = 0 #背景序號

detector = handDetector() #建立偵測class

catched=False

garbage=[]

food=[]

random.seed(a=None,version=2)

gameset=False

gamestart=False

pathFullBackground = os.path.join(backgroundPath, pathBackgrounds[backgroundNumber])

imgBackground = cv2.imread(pathFullBackground)

imgGarbage = cv2.imread("assets/P_garbage.png", cv2.IMREAD_UNCHANGED) 
imgGarbage = cv2.resize(imgGarbage, (0,0), None, 0.25, 0.25)

imgFood = cv2.imread("assets/P_food.png", cv2.IMREAD_UNCHANGED)
imgFood = cv2.resize(imgFood, (0,0), None, 0.25, 0.25)

imgCat = cv2.imread("assets/P_cat.png", cv2.IMREAD_UNCHANGED)
imgCat = cv2.resize(imgCat, (0,0), None, 0.25, 0.25)

catx=5
caty=5

overlap=0

end=1

while True:
    
    if end==1:
        garbage=[]
        end=0
    
    if gameset==False:
        for i in range(4):
            x=random.randint(100,540 ) #垃圾位置的x軸
            y=random.randint(50,290 ) #垃圾位置的y軸
            garbage.append([i,x,y])
            
            
        while overlap==0:
            x=random.randint(100,540 ) #食物位置的x軸
            y=random.randint(50,290 ) #食物位置的y軸
            overlap=1
            food=[x,y]
            for i in range(4):
                if x==garbage[i][1]:
                    if y==garbage[i][2]:
                        overlap=0
        gameset=True 
        
    
    retval, img = cap.read() #讀取攝像頭照到的影像，retval表示影像捕獲成功or失敗
    
    img = detector.findHands(img) #偵測手掌
    
    landmarkList = detector.findPosition(img, 0) #獲取手掌偵測節點資料
    
    img = np.flip(img, axis=1) #攝像頭畫面左右翻轉(axis=1表示左右翻轉，axis=0表示上下翻轉)
    img = np.ascontiguousarray(img)
    
    #pathFullBackground = os.path.join(backgroundPath, pathBackgrounds[backgroundNumber]) #存放在背景圖片資料夾(backgroundPath)的第一個背景
    #imgBackground = cv2.imread(pathFullBackground) #讀取該背景圖片
    pathFullBackground = os.path.join(backgroundPath, pathBackgrounds[backgroundNumber])
    imgBackground = cv2.imread(pathFullBackground)
    
    gameSize_x, gameSize_y = imgBackground.shape[1], imgBackground.shape[0] #獲取背景圖片大小(作為遊玩畫面大小)
    offset_x, offset_y = int((cam_width - gameSize_x) / 2), int((cam_height - gameSize_y) / 2)
    limitMin_x, limitMin_y = offset_x, offset_y
    limitMax_x, limitMax_y = cam_width-offset_x, cam_height-offset_y 
    cv2.rectangle(img, (limitMin_x, limitMin_y), (limitMax_x, limitMax_y), (255,0,0), 2)
    
    if len(landmarkList) != 0:
        finger_x, finger_y = -(landmarkList[8][1] - cam_width), landmarkList[8][2] #獲取食指座標
        gamePoint = (finger_x - limitMin_x, finger_y - limitMin_y)
        
        if finger_x > limitMin_x and finger_x < limitMax_x and finger_y > limitMin_y and finger_y < limitMax_y:
            cv2.circle(imgBackground, gamePoint, 10, (0, 0, 255), cv2.FILLED) 
        
        if gamePoint[0]>=catx-5 and gamePoint[0]<=catx+5:
            if gamePoint[1]>=caty-5 and gamePoint[1]<=caty+5:
                gamestart=1
                print("susss")
        if gamestart==1:
            print(gamePoint[0])
            print("-----")
            print(gamePoint[1])
            for i in range(4):
                if gamePoint[0]>=(garbage[i][1])-5 and gamePoint[0]<=(garbage[i][1])+5:
                    if gamePoint[1]>=(garbage[i][2])-3 and gamePoint[1]<=(garbage[i][2])+3:
                        print("eisisiw")
                        gameset=False
                        end=1
                        
    imgBackground = cvzone.overlayPNG(imgBackground, imgCat, (catx, caty))
    imgBackground = cvzone.overlayPNG(imgBackground, imgFood, (food[0], food[1]))
    for i in range(4):
        imgBackground = cvzone.overlayPNG(imgBackground, imgGarbage, (garbage[i][1], garbage[i][2]))
    
    
    cv2.namedWindow("Image", cv2.WINDOW_NORMAL) #新增攝像頭視窗(Image)，允許視窗自由縮放
    cv2.resizeWindow("Image", img.shape[1]//2, img.shape[0]//2) #改變"Image"視窗大小
    cv2.imshow("Image", img) #攝像頭照到的影像顯示到"Image"視窗
    
    cv2.namedWindow("Game", cv2.WINDOW_NORMAL) #新增遊戲視窗(Game)，允許視窗自由縮放
    cv2.resizeWindow("Game", imgBackground.shape[1]*2, imgBackground.shape[0]*2) #改變"Game"視窗大小
    cv2.imshow("Game", imgBackground) #將背景圖片顯示到"Game"視窗
    
    key = cv2.waitKey(1) #等待1毫秒接收按鍵
    if key == 27: #若按下Esc按鍵
        break #退出迴圈
    
cap.release() #結束攝像機

cv2.destroyAllWindows() #關閉所有視窗
```
